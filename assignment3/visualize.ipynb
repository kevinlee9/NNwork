```python
import numpy as np
from matplotlib import pyplot as plt 
import random
```

```python
# Original Image
nx = 10  # figure with 10x10 digits
ny = 5
digit_size = 28
# data_tr = np.load("data.np")
# data_lb = np.asarray(np.load("lbs.np"))
figure = np.zeros((digit_size * ny, digit_size * nx ))
# i for digit type
for i in range(0, nx):
    index_class = data_lb == i
    class_idx, = np.where(data_lb == i)
    sample_data = data_tr[class_idx]
    # j for number of sampling
    for j in range(0, ny):
        digit_ori = sample_data[j].reshape(digit_size, digit_size)
        figure[j * digit_size: (j + 1) * digit_size,
        i * digit_size: (i + 1) * digit_size] = digit_ori


plt.figure(figsize = (nx, ny))
plt.imshow(figure, cmap = 'Greys_r')
frame = plt.gca()
frame.axes.get_yaxis().set_visible(False)
frame.axes.get_xaxis().set_visible(False)
plt.show()
plt.savefig('dataset.pdf')
```

```python
# before training
C1 = np.load("h_conv1_i.np")
C3 = np.load("h_pool1_i.np")
C5 = np.load("h_pool2_i.np")
data_tr = np.load("data.np")
data_lb = np.asarray(np.load("lbs.np"))
title = ['orignial','C1 feature map','C3 feature map','C5 feature map']
fig, axes = plt.subplots(10, 4, figsize=(9, 24),
                         subplot_kw={'xticks': [], 'yticks': []})
fig.subplots_adjust(hspace=0.1, wspace=0.05)
for i, ax in enumerate(axes.flat):
    
    if i%4 == 0:
        class_idx, = np.where(data_lb == i//4)
        index = class_idx[0]
        original_data = data_tr[index]
#         C1, C3, C5 = lenet.get_deep_features(sess, original_data)
        
    if i%4 == 0:
        ax.imshow(original_data.reshape(28,28), cmap='Greys_r')
    if i%4 == 1:
        ax.imshow(np.mean(C1,axis=3)[index], cmap='Greys_r')
    if i%4 == 2:
        ax.imshow(np.mean(C3,axis=3)[index], cmap='Greys_r')
    if i%4 == 3:
        ax.imshow(np.mean(C5,axis=3)[index], cmap='Greys_r')
    if i <=3:
        ax.set_title(title[i])
plt.savefig("LeNet feature map before training.pdf")
plt.show()
```
